```python
import csv
import matplotlib.pyplot as plt
# 年份和GDP數據
years = list(range(2014, 2024))
taiwan_gdp = [530.5, 523.6, 529.5, 571.0, 602.0, 610.8, 586.2, 668.1, 720.0, 760.4]  # 臺灣的GDP數據(億美元)
japan_gdp = [4846.9, 4389.5, 4926.6, 4872.4, 4950.9, 5081.8, 5064.8, 5151.0, 4240.0, 4210.6]  # 日本的GDP數據(億美元)
usa_gdp = [17608.1, 18295, 18804.9, 19612.1, 20656.5, 21521.4, 21323, 23594, 25744.1, 27356.4 ]  # 美國GDP數據(億美元)
Thailand_gdp = [3660, 4127, 5063.30, 4952.10, 5059.8, 5025.8, 5059.82, 5040.82, 5124.43, 5124.43 ]  # 泰國GDP數據(億美元)

# 將數據寫入CSV檔案
with open('gdp_data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Year", "Taiwan GDP", "Japan GDP", "USA GDP", "Thailand GDP"])
    for i in range(len(years)):
        writer.writerow([years[i], taiwan_gdp[i], japan_gdp[i], usa_gdp[i], Thailand_gdp[i]])
```

```python
years = list(range(2014, 2024))
```

```python
# 繪製GDP折線圖
plt.figure(figsize=(10, 5))

plt.plot(years, taiwan_gdp, marker='o', label='Taiwan')
plt.plot(years, japan_gdp, marker='o', label='Japan')
plt.plot(years, Thailand_gdp, marker='o', label='THailand')
plt.plot(years, usa_gdp, marker='o', label='USA')


plt.xlabel('years')
plt.ylabel('GDP (billion US dollars)')
plt.title('Comparison of GDP between Taiwan, Japan, the United States and Thailand from 2014 to 2023')
plt.legend()

plt.show()
```
